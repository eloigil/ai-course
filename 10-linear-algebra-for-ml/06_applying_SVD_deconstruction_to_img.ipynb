```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```
plt.style.use('classic')

img = Image.open('imagen_ejemplo_frida_bredesen.jpg')

plt.imshow(img)
```

```
img_gray = img.convert('LA')

img_matrix = np.array(list(img_gray.getdata(band=0)), float)

img_matrix.shape = (img_gray.size[1], img_gray.size[0])
```

```
U,D,V = np.linalg.svd(img_matrix)
```

```
print(img_matrix.shape)
print(U.shape)
print(D.shape)
print(V.shape)
```

```
img_reconstruction = np.matrix(U[:,:1]) * np.diag(D[:1]) * np.matrix(V[:1,:])
plt.imshow(img_reconstruction, cmap='gray')
```
