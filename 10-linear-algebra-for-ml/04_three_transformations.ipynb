```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```
def graphVectors(vects, cols, alpha=1):
  plt.axvline(x=0, color='grey', zorder=0)
  plt.axhline(y=0, color='grey', zorder=0)

  for i in range(len(vects)):
    x = np.concatenate([[0,0], vects[i]])
    plt.quiver([x[0]],
               [x[1]],
               [x[2]],
               [x[3]],
               angles='xy',
               scale_units='xy',
               scale=1,
               color=cols[i],
               alpha=alpha
              )
```

```
def graphMatrix(matrix, vectorColors=['red', 'blue']):
  # unitary circle
  x = np.linspace(-1,1,100000)
  y = np.sqrt(1-(x**2))

  # transformed unitary circle
  x1 = matrix[0,0]*x + matrix[0,1]*y
  y1 = matrix[1,0]*x + matrix[1,1]*y
  x1_neg = matrix[0,0]*x - matrix[0,1]*y
  y1_neg = matrix[1,0]*x - matrix[1,1]*y

  # vectors
  u1 = [matrix[0,0], matrix[1,0]]
  v1 = [matrix[0,1], matrix[1,1]]

  graphVectors([u1, v1], cols=[vectorColors[0], vectorColors[1]])

  plt.plot(x1,y1, 'green', alpha=0.7)
  plt.plot(x1_neg, y1_neg, 'green', alpha=0.7)
```

```
A = np.array([[3,7],[5,2]])
print(A)
```

```
print('unitary circle')
graphMatrix(np.array([[1,0],[0,1]]))

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()

print('transformed unitary circle')
graphMatrix(A)

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```

## Applications and effects of UDV matrices


```
U,D,V = np.linalg.svd(A)
print(U)
```

```
print('unitary circle')
graphMatrix(np.array([[1,0],[0,1]]))

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()

print('First rotation V:')
graphMatrix(V)

plt.xlim(-1.5,1.5)
plt.ylim(-1.5,1.5)
plt.show()

print('Scale D:')
graphMatrix(np.diag(D).dot(V))

plt.xlim(-9,9)
plt.ylim(-9,9)
plt.show()

print('Second rotation U:')
graphMatrix(U.dot(np.diag(D).dot(V)))

plt.xlim(-8,8)
plt.ylim(-8,8)
plt.show()
```
