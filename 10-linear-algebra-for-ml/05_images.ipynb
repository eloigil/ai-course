```
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```
plt.style.use('classic')

img = Image.open('imagen_ejemplo_frida_bredesen.jpg')

plt.imshow(img)
```

```
img_gray = img.convert('LA')
print(img_gray)
```

```
img_matrix = np.array(list(img_gray.getdata(band=0)), float)
print(img_matrix)
```

```
img_matrix.shape = (img_gray.size[1], img_gray.size[0])
print(img_matrix)
print(img_matrix.shape)
```

```
plt.imshow(img_matrix, cmap='gray')
plt.show()
```

```
img_matrix2 = img_matrix / 10
print(img_matrix2)
```

```
plt.imshow(img_matrix2)
plt.show()
```

```
plt.imshow(img_matrix2, cmap='gray')
plt.show()
```

```
print(np.max(img_matrix2))
print(np.max(img_matrix))
```

```
print(np.min(img_matrix2))
print(np.min(img_matrix))
```
